```python
print(np.arange(10))
print(np.arange(1, 10))
print(np.arange(1, 10, .7))
print(np.arange(10, 1, -.8))
print(np.arange(10, 0, -.9))
```

```python
print(np.arange(3) + np.arange(4, 7))
print(np.arange(1, 10) * 2)
print(np.arange(3) - np.arange(3))
print(np.arange(3) * np.arange(3))
print(np.arange(3) / np.arange(1, 4))
```

```python
import matplotlib.pyplot as plt
import random

x = []
y = []
size = []

for i in range(100) : 
    x.append(random.randint(50,100))
    y.append(random.randint(50,100))
    size.append(random.randint(10,200))
    
plt.scatter(x, y, s=size, c=size, cmap='jet')
plt.grid(True)
plt.colorbar()
plt.show()
```
